```python
import pandas as pd
purchase_behav=pd.read_csv("purchase_behaviour.csv")
transaction_data=pd.read_csv('transaction_data.csv')
```

```python
pd.set_option("display.max_rows", None)

pd.set_option("display.max_columns", None)

transaction_data.shape
```

```python
transaction_data.head(11)
```

```python
values = transaction_data["DATE"].value_counts().sort_index(ascending=True)
print(values)
```

```python




top_customers = transaction_data.groupby("LYLTY_CARD_NBR")["TOT_SALES"].sum().sort_values(ascending=False).head(200)


print("\n Top 10 Buying Customers:")
print(top_customers.tail(50))

```

```python

customer_226000 = transaction_data[transaction_data["LYLTY_CARD_NBR"] == 63197 ]


total_days = customer_226000["DATE"]
total_days.shape
```

```python



customer_purchase_days = transaction_data.groupby("LYLTY_CARD_NBR")["DATE"].count().sort_values(ascending=False).head(200)


print(customer_purchase_days)
```

```python
common_products =customer_purchase_days.index.intersection(top_customers.index)
len(common_products)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


top_50_customers = top_customers.head(50)

plt.figure(figsize=(12, 6))


sns.barplot(x=top_50_customers.index, y=top_50_customers.values, palette="viridis")


plt.xticks(rotation=90)


plt.xlabel("Loyalty Card Number")
plt.ylabel("Total Sales")
plt.title(" Top 10 Buying Customers (Total Sales)")


plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


top_50_purchase_days = customer_purchase_days.head(10)


plt.figure(figsize=(12, 6))


sns.barplot(x=top_50_purchase_days.index, y=top_50_purchase_days.values, palette="coolwarm")


plt.xticks(rotation=90)

plt.xlabel("Loyalty Card Number")
plt.ylabel("Total Purchase Days")
plt.title(" Top 10 Customers by Total Purchase Days in One Year")


plt.show()
```
